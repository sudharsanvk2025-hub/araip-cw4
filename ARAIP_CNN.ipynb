```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing import image

# 1. Load CIFAR-10 Dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
class_names = ['deer', 'automobile', 'bird', 'cat', 'airplane', 'dog', 'frog', 'horse', 'ship', 'truck']

# 2. Data Augmentation (Optimization Step)
# This helps the model generalize so birds aren't mistaken for trucks
data_augmentation = tf.keras.Sequential([
  layers.RandomFlip("horizontal"),
  layers.RandomRotation(0.1),
  layers.RandomZoom(0.1),
])
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# 1. Dataset Preparation (CIFAR-10)
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

# Data Augmentation Layer
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
])

# 2. Define the Model Architecture (As specified in your requirement)
def build_enhanced_model():
    model = models.Sequential([
        data_augmentation, # Apply augmentation to training

        # Block 1: Initial feature extraction
        layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),

        # Block 2: Deeper features
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.3),

        # Block 3: Complex patterns
        layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(10)
    ])
    return model

# 3. Training with ADAM
print("Starting Training with ADAM Optimizer...")
model_adam = build_enhanced_model()
model_adam.compile(optimizer='adam',
                   loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                   metrics=['accuracy'])

history_adam = model_adam.fit(train_images, train_labels, epochs=10,
                              validation_data=(test_images, test_labels))

# 4. Training with SGD
print("\nStarting Training with SGD Optimizer...")
model_sgd = build_enhanced_model()
model_sgd.compile(optimizer='sgd',
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])

history_sgd = model_sgd.fit(train_images, train_labels, epochs=10,
                             validation_data=(test_images, test_labels))

# 5. Generate Comparison Graphs
plt.figure(figsize=(14, 6))

# Graph 1: Accuracy Comparison
plt.subplot(1, 2, 1)
plt.plot(history_adam.history['accuracy'], label='Adam Train Acc', color='blue', linestyle='--')
plt.plot(history_adam.history['val_accuracy'], label='Adam Val Acc', color='blue')
plt.plot(history_sgd.history['accuracy'], label='SGD Train Acc', color='red', linestyle='--')
plt.plot(history_sgd.history['val_accuracy'], label='SGD Val Acc', color='red')
plt.title('Training and Validation Accuracy (Adam vs SGD)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Graph 2: Loss Comparison
plt.subplot(1, 2, 2)
plt.plot(history_adam.history['loss'], label='Adam Train Loss', color='blue', linestyle='--')
plt.plot(history_adam.history['val_loss'], label='Adam Val Loss', color='blue')
plt.plot(history_sgd.history['loss'], label='SGD Train Loss', color='red', linestyle='--')
plt.plot(history_sgd.history['val_loss'], label='SGD Val Loss', color='red')
plt.title('Training and Validation Loss (Adam vs SGD)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# 6. Show Model Summary (Layers) [cite: 7, 17]
model_adam.summary()
```
